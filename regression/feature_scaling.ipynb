```python
import numpy as np
import pandas as pd
from random import uniform
import matplotlib.pyplot as plt
import pylatex
```

```python
df = pd.DataFrame({
    "meters":[ 67,  74,  93,  96, 102, 114, 127,  182, 193],
    "rooms": [  2,   3,   3,   3,   4,   3,   3,    5,   4],
    "prices":[380, 450, 630, 640, 730, 720, 850, 1020, 980]})
```

```python
def make_graphs(d):
    plt.figure(figsize=(16, 5))
    plt.subplot2grid((1, 2), (0, 0))
    plt.scatter(d["meters"], d["prices"])
    plt.xlabel("meters")
    plt.ylabel("prices")
    plt.subplot2grid((1, 2), (0, 1))
    plt.scatter(d["rooms"], d["prices"])
    plt.xlabel("rooms")
    plt.ylabel("prices")
```

# Correlation

```python
print(df.corr())
```

```python
make_graphs(df)
```

# Feature Scaling

```python
def create_new_df(fun):
    return pd.DataFrame({
        "meters":fun("meters"),
        "rooms": fun("rooms"),
        "prices":fun("prices")
        })
```

## Mean Normalization

$$x' = \frac{x - average(x)}{max(x) - min(x)}$$

```python
def subtract_mean(f):
    return df[f] - df[f].mean()
```

```python
make_graphs(create_new_df(subtract_mean))
# data is zero-centered
```

```python
def mean_scale(f):
    return (subtract_mean(f)) / (df[f].max() - df[f].min())
```

```python
make_graphs(create_new_df(mean_scale))
# after mean normalization
```

## Standardization

$$x' = \frac{x - \bar{x}}{\sigma}$$

```python
def standard_scale(f):
    return (df[f] - df[f].mean()) / df[f].std()
```

```python
make_graphs(create_new_df(standard_scale))
```
