# Linear Regression First

```python
import matplotlib.pyplot as plt
import pylatex
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from random import uniform
import math
```

```python
df = pd.DataFrame({"meters":[ 67,  74,  93,  96, 102, 114, 127,  182, 193],
                   "price": [380, 450, 630, 640, 730, 720, 850, 1020, 980]})
```

```python
X_train = np.array(df["meters"]).reshape(-1,1)
y_train = np.array(df["price"])
```

```python
X_train
```

```python
y_train
```

### Sklearn

```python
# fit_intercept indica si queremos usar el par√°metro n o no
clf = LinearRegression(fit_intercept=False)
clf.fit(X_train, y_train)
y_test = clf.predict(X_train)
print(f"m = {clf.coef_[0]}")
print(f"n = {clf.intercept_}")
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df["meters"], y_train, "bo", 
         df["meters"], y_test, "r")
```

```python
clf = LinearRegression(fit_intercept=True)
clf.fit(X_train, y_train)
y_test = clf.predict(X_train)
print(f"m = {clf.coef_[0]}")
print(f"n = {clf.intercept_}")
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df["meters"], y_train, "bo", 
         df["meters"], y_test, "r")
```
